# Carga de dataset

```python
# Cargar el dataset desde Google Drive
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/MNA/Operaciones de aprendizaje automático/Fase 1/v1/Residential-Building-Data-Set.xlsx')
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
df.head()
```

# Manipulación y preparación de datos

```python
# Cambiar nombres de columnas para evitar duplicados
column_names = []
for i in range(len(df.columns)):
    column_names.append(df.columns[i])

for i in range(len(df.columns)):
    if df.columns[i] in column_names:
        column_names[i] = column_names[i] + '_' + str(i)

df.columns = column_names
```

```python
# Convertir todas las columnas a numéricas (si es posible)
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

# EDA

```python
# Matriz de correlación para observar las variables
sns.set(rc={'figure.figsize': (20, 15)})  # Ajustar tamaño de la figura
matrix = df.corr(numeric_only=True)
sns.heatmap(matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Función para contar columnas con valores faltantes
def contador_valores_faltantes(df):
    missing_count = df.isnull().sum()
    missing_columns = missing_count[missing_count > 0]
    print(f"Numero de columnas con valores faltantes: {len(missing_columns)}")
    print(missing_columns)
```

```python

# Llamar la función para contar valores faltantes
contador_valores_faltantes(df)
```

```python
# Función para contar columnas con valores únicos
def count_single_unique_value_columns(df):
    return sum(df.nunique() == 1)
```

```python

# Llamar la función para contar columnas con valores únicos
result = count_single_unique_value_columns(df)
print(f"Numero de columnas con valores unicos: {result}")
```

# Exploración y preprocesamiento de datos

```python
#Escalado de datos
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df.select_dtypes(include=[np.number])), columns=df.select_dtypes(include=[np.number]).columns)
df_scaled.head()
```

```python
# Reducción de dimensionalidad con PCA
pca = PCA(n_components=2)  # Reducir a 2 componentes principales
pca_data = pca.fit_transform(df_scaled)
pca_df = pd.DataFrame(data=pca_data, columns=['PCA1', 'PCA2'])
pca_df.head()
```

```python
# Visualizar los resultados de la PCA
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c='blue', marker='o')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('Reducción de dimensionalidad con PCA')
plt.show()
```
